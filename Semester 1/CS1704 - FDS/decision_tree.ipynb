# Decision Tree

Decision trees are a supervised learning model used for classification and regression in order to predict the target variables.
<br>

The main objective of a decision tree is to predict the target variable by making simple decisions from the data features.

To plot decision trees, we are going to import `DecisionTreeClassifier` and `plot_tree` from `sklearn.tree`.

```python
# Import necessary libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Load the iris dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # Target labels

# Create a decision tree classifier
clf = DecisionTreeClassifier(max_depth=3, random_state=42)

# Fit the classifier to the data
clf.fit(X, y)

# Visualize the decision tree
plt.figure(figsize=(12, 8))
plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.show()
```

In the given code, we imported the iris dataset and separated the data and target values in different variables. Then, we fit our dataset in the Decision Tree model, and finally we create an image of the tree and display it.
<br>

This model splits the dataset into subsets based on the features in a tree-like structure. Each internal node corresponds to a feature, each branch talks about a decision and each leaf node gives the final outcome of the class.

## Decision Tree Classifier

`clf = DecisionTreeClassifier(max_depth=3, random_state=42)`
<br>

If you notice, we passed in two parameters to this model:
- `max_depth`, and
- `random_state`

The `max_depth` parameter defines how deep our tree should go or how many times we want the model to split data to make decisions.
<br>

The `random_state` parameter defines how consistent the model is supposed to be when the code is run each time. This provides reproducibility.

## `plot_tree` Function

`plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)`

We have passed four parameters to this function:
- `DecisionTreeClassifier` object which is `clf` in this case
- `feature_names` where we pass the list of features in the data
- `class_names` where pass the list of class names in the data
- `filled` which colours the nodes of the tree, making it easy to visualize

## Decision Tree Construct

- **Splitting**: At each node, the algorithm chooses the best feature and threshold to split the dataset based on a criterion like Gini impurity or entropy (Information Gain).
- **Gini Impurity**: Measures how often a randomly chosen element from the set would be incorrectly classified.
- **Entropy**: Measures the uncertainty or disorder in the dataset. The goal is to split the data in a way that maximizes homogeneity within the resulting subsets.
- **Recursion**: After splitting, the algorithm recursively repeats the process for each child node until a stopping condition is met like `max_depth`, `min_samples_split`, etc.
- **Leaf Nodes**: When a stopping criterion is reached, the tree stops splitting, and the final predictions (classes) are assigned to the leaf nodes.


## Decision Tree Classifier: Advantages vs Disadvantages

| **Advantages** | **Disadvantages** |
|----------------|-------------------|
| Easy to understand and interpret (similar to human decision-making) | Prone to **overfitting**, especially with noisy or small datasets |
| Requires little preprocessing (no scaling or normalization needed) | Highly sensitive to small changes in data (high variance) |
| Handles both numerical and categorical data | Trees can grow very deep, becoming complex and hard to generalize |
| Works for both classification and regression tasks | Biased toward features with many categories/levels |
| Naturally handles non-linear relationships | Requires pruning or parameter tuning to avoid overfitting |
| Fast to train and predict | Greedy split decisions may not produce the optimal tree |
| Handles missing values effectively | Not ideal for high-dimensional datasets without constraints |
| Gives feature importance scores | A single tree usually performs worse than ensemble methods |
| Works very well when combined with ensemble models like Random Forest | Unstable unless hyperparameters like `max_depth` are controlled |
