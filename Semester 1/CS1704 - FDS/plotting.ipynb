# Plotting

When it comes to data science, plotting data becomes a crucial part of building an efficient model because we are dealing with data, and data goes very well with plotting.
<br>

It is important to plot data because it provides a lot of useful insights from the data irrespective of how large it is. Moreover, our models may use it to process data efficiently.
<br>

There are various ways to plot data:
- Box Plot
- Scatter Plot
- Kernel Density Estimate (KDE)
- Violin Plot
- Histogram
- Bar Plot
- Pair Plot, and more
<br>

In this PDF, we are going look at three kinds of plots. To start plotting use the `seaborn` library which provides a wide variety of data plotting techniques, and `matplotlib.plyplot` to plot the graphs.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Load a dataset for plotting graphs

dataset = load_iris()
df = pd.DataFrame(data=dataset.data, columns=dataset.feature_names)
```

## Box Plot

A box plot, also called a box-and-whisker plot is used to show the distribution of quantitative data such that it facilitates a way to compare categorical variables.
<br>

It has a box which shows the quartiles and the whiskers which extend to the minimum and maximum values in the data. The points on the graph represent "outliers".
<br>

Outliers are those data points which are meaningless in the data and impact the outcome of the data. It must be removed for a better acccuracy.
<br>

The lower part of the box represents Quartile 1, or the point where 25% of data lie below it. The upper part represents Quartile 3, or the point where 75% of data lies below it. There is a line in between the box. This represents Quartile 2, or the point where 50% of the data lies below it.
<br>

The range in between Q1 and Q3 (The box) is called the **inter quartile range** and it can be found by:
$$IQR = Q_1 + Q_3$$
<br>

To plot a box plot, use the `boxplot` function from `seaborn`.

```python
sns.boxplot(data=df)
plt.show()
```

## Kernel Density Estimate (KDE)

A Kernel Density Estimate (KDE) is used to draw a smooth distribution estimate for a given data. It works like a histogram, except the fact that histograms have connected rectangles, and this is a continuous curve that is smooth.
<br>

Unlike histograms, which requires bins and the graph varies based on the number of bins chosen, KDE is similar in all cases. It does so by creating multiple bell shaped curves, called **kernel**.
<br>

Reading a KDE is simple. It works  just like a histogram. There are lots of data at places where the curve peaks and little data where the curve creates a valley.
<br>

To plot KDE, use the `kdeplot` function from `seaborn`.

```python
sns.kdeplot(data=df)
plt.show()
```

## Violin Plot

A violin plot is combination of the box plot and the KDE. The plot is shaped like a violin, and hence the name of this plot.
<br>

The KDE plots a smooth curve of the given data, describing the borders of the violin plot. The box plot lies at the centre of the plot along the plot, describing the interquartile range and outliers.
<br>

It behaves similar to a box plot, except it contains a distribution which helps in comparing multiple categorical variables. This makes it better than the box plot because it shows both the quartiles and the distribution of the data in an attractive way.
<br>

To plot a violin plot, use the `violinplot` function from `seaborn`.

```python
sns.violinplot(data=df)
plt.show()
```

# Scatter Plot

The scatter plot is used to plot data points on a cartesian plane that has two features on the axes. It can be great to visualize the density of data at some region or condition and a grouping data based on a certain criteria.
<br>

To make a scatter plot, use the `scatter plot` function from `seaborn`. A basic plot needs these three parameters:
- `x` is the feature on the `x-axis`
- `y` is the feature on the `y-axis`
- `data` is the data to be plotted.

To access a feature of a data, use the syntax `data_file[feature_name]` or indexing.

```python
sns.scatterplot(x=df.columns[0], y=df.columns[1], data=df)
plt.show()
```

# Bar Plot

Seaborn's `barplot` shows a central tendency (by default the mean) of a numeric variable for each level of a categorical variable, with error bars to indicate uncertainty (confidence intervals).

Key points
- Purpose: visualize aggregated summary (e.g., mean) of a numeric variable per category.
- Default behavior: estimator = mean, ci = 95 (95% confidence interval shown as error bars).
- Use when you want to compare summary statistics across categories. For raw counts use `sns.countplot`.

Common parameters
- `x`, `y`: column names or arrays (one should be categorical, the other numeric).
- `hue`: group within each category (creates colored bars).
- `data`: DataFrame.
- `estimator`: function to compute the central value (e.g., `np.mean`, `np.median`).
- `ci`: confidence interval to draw (e.g., 95) or `None` to disable.
- `order`, `hue_order`: control category ordering.
- `palette`: color palette.
- `capsize`: width of error bar caps (float).
- `dodge`: separate bars for `hue` levels (True/False).
- `orient`: 'v' or 'h' to force vertical/horizontal orientation.

```python
sns.barplot(x=df.columns[0], y=df.columns[1], data=df)
plt.xticks(rotation=90) # Rotate ticks (labels) on the x-axis
plt.show()
```
